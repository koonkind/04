### Read Dataset (csv)

```python
url='https://raw.githubusercontent.com/worachaiSM/tutorial/main/mpg.csv'
df=pd.read_csv(url)
df.head(20)
```

# plot แบบ wide format

```python
sns.barplot(data=[df.cty, df.hwy]);
```

```python
sns.barplot(data=df[["cty","hwy"]]);
```

```python
sns.barplot(data=df[['cty','hwy']], orient='h');
```

# Plot Long Format

```python
sns.barplot(data=df, x='cty');
```

```python
sns.barplot(data=df, y='cty');
```

```python
df['gear']=df.trans.str[0].map({'a':'auto', 'm':'manual'}).astype("category")
df['drv']=df.drv.map({'f':'front', 'r':'rear', '4':'4-wheel'}).astype("category")
df.head(20)
```

```python
sns.barplot(data=df, x='gear', y='cty', palette='winter');
```

```python
sns.barplot(data=df, x='gear', y='cty',
            hue='drv', palette='twilight_r');
```

```python
sns.barplot(data=df, x='gear', y='hwy',
            hue='drv', palette='inferno_r');
```

# Factor plot

```python
sns.barplot(data=df, x='drv', y='cty',
              hue='gear');
```

```python
sns.factorplot(data=df, x='drv', y='cty',
              hue='gear',
              col='year',
              kind='bar',
              palette='twilight');
```

```python
sns.factorplot(data=df, x='drv', y='cty',
               hue='gear',
               row='class', col='year',
               kind='bar');
```
